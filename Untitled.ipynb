```python
#reading in our Airbnb Data and adding it to our Dataframe
df = pd.read_csv(r'C:\Users\Justin\Documents\GitHub\Web_Scrape\AirDNA_Surrey.csv',encoding='ISO-8859-1')
df.head(3)
```

```python
#Create a new dataframe containing all of our homes that only have 2 bedrooms
df2 = df.query('bedrooms == 2')
df2. head()
```

```python
#Checking the file information
df2.info()
```

```python
#creating a list with our column headers
column_headers = list(df2.columns.values)
#print(column_headers)
#creating a list that contains the columns headers we will be removing
columns_to_remove = column_headers[-4:]
#print(columns_to_remove)
#using pandas built in to remove columns
df2.drop(columns_to_remove, axis=1, inplace=True)
df2.head()
```

```python
#Checking the file information to find issues with our data
df2.info()
```

It appears as though we will need to switch certain columns from a string to a numeric type.
The following fields will need to be changed: rating and bathrooms

```python
#finding our unique values in the rating
df2.rating.unique()

#replacing all null strings to 0
df2.rating = df2.rating.replace("null", 0)

#checking our work
df2.rating.unique()

#converting column to numeric
df2[["rating"]] = df2[["rating"]].apply(pd.to_numeric)

#checking our unique values once again
df2.rating.unique()
```

```python
#We will now convert all our values in bathroom to a numeric value
#finding our unique values in the rating
unique_bath = df2.bathrooms.unique()
#print(unique_bath)

#converting our values from string
df2[["bathrooms"]] = df2[["bathrooms"]].apply(pd.to_numeric)

#checking our work
unique_bath = df2.bathrooms.unique()
print(unique_bath)
```

We will further shrink our data set to ensure that we are only keeping columns that will bring us value
We will be removing the following columns, airbnb_property_id, homeaway_property_id, m_homeaway_property_id, id, room_type, platforms__airbnb_property_id, and listings_returned.

```python
#create a list of columns to remove
columns_to_remove=['airbnb_property_id', 'homeaway_property_id', 'm_homeaway_property_id', 'platforms__airbnb_property_id', 'platforms__homeaway_property_id', 'listings_returned','room_type']
columns_to_remove=['id']
#Removing unnecessary columns
df2 = df2.drop(columns_to_remove, axis=1)
df2.head()
```

```python
#We will be using linear regression to create a model
import statsmodels.api as sm

#I will begin by adding an intercept
df2['intercept'] = 1
```

```python
#We will be running a linear regression model on our df to find the relevant pieces in determining revenue
lm = sm.OLS(df2['revenue'], df2[['rating', 'reviews', 'bathrooms']])
results = lm.fit()
results.summary()
```
