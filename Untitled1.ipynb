```python
import requests
import pandas as pd
import bs4


# Get all prices for phones on sale on craigslist
r = requests.get("https://vancouver.craigslist.org/search/sss?query=phone&sort=rel")
exampleSoup = bs4.BeautifulSoup(r.text, 'html.parser')

price_elem = exampleSoup.find_all('span', {'class':"result-price"})

for i in price_elem:
    print(i.text)

```
